### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

```python
import matplotlib.pyplot as plt
```

#1. Análise do faturamento




```python
print(loja.head())
```

```python
loja['Faturamento'] = loja['Preço'] * loja['Quantidade de parcelas']
loja2['Faturamento'] = loja2['Preço'] * loja2['Quantidade de parcelas']
loja3['Faturamento'] = loja3['Preço'] * loja3['Quantidade de parcelas']
loja4['Faturamento'] = loja4['Preço'] * loja4['Quantidade de parcelas']
```

```python
faturamento_loja1 = loja['Faturamento'].sum()
faturamento_loja2 = loja2['Faturamento'].sum()
faturamento_loja3 = loja3['Faturamento'].sum()
faturamento_loja4 = loja4['Faturamento'].sum()

print(f"Faturamento Loja 1: R$ {faturamento_loja1:.2f}")
print(f"Faturamento Loja 2: R$ {faturamento_loja2:.2f}")
print(f"Faturamento Loja 3: R$ {faturamento_loja3:.2f}")
print(f"Faturamento Loja 4: R$ {faturamento_loja4:.2f}")
```

```python
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
```

```python
plt.figure(figsize=(8, 6))
plt.bar(lojas, faturamentos, color='skyblue')

plt.title('Faturamento Total por Loja')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

# 3. Média de Avaliação das Lojas

```python
def media_avaliacao(df):
    return df['Avaliação da compra'].mean()

medias = {
    'Loja 1': media_avaliacao(loja),
    'Loja 2': media_avaliacao(loja2),
    'Loja 3': media_avaliacao(loja3),
    'Loja 4': media_avaliacao(loja4)
}

for loja, media in medias.items():
    print(f'{loja}: {media:.2f}')
```

```python
lojas = list(medias.keys())
medias_avaliacoes = list(medias.values())

plt.figure(figsize=(8,5))
plt.bar(lojas, medias_avaliacoes, color='skyblue')
plt.title('Média de Avaliação por Loja')
plt.xlabel('Lojas')
plt.ylabel('Média de Avaliação')
plt.ylim(0, 5)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, v in enumerate(medias_avaliacoes):
    plt.text(i, v + 0.1, f'{v:.2f}', ha='center', fontweight='bold')

plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
def produtos_mais_menos_vendidos(df, nome_loja):
    contagem = df['Produto'].value_counts()

    mais_vendido = contagem.idxmax()
    menos_vendido = contagem.idxmin()

    print(f'📦 {nome_loja}')
    print(f'Produto mais vendido: {mais_vendido} ({contagem.max()} vendas)')
    print(f'Produto menos vendido: {menos_vendido} ({contagem.min()} vendas)\n')

# Analisar cada loja
produtos_mais_menos_vendidos(loja, 'Loja 1')
produtos_mais_menos_vendidos(loja2, 'Loja 2')
produtos_mais_menos_vendidos(loja3, 'Loja 3')
produtos_mais_menos_vendidos(loja4, 'Loja 4')
```

```python
# Função para plotar o Top 5 produtos mais vendidos
def top5_produtos_mais_vendidos(df, nome_loja):
    top5 = df['Produto'].value_counts().head(5)

    plt.figure(figsize=(8,5))
    plt.barh(top5.index, top5.values, color='mediumseagreen')
    plt.title(f'Top 5 Produtos Mais Vendidos - {nome_loja}')
    plt.xlabel('Quantidade Vendida')
    plt.ylabel('Produto')
    plt.gca().invert_yaxis()
    for i, v in enumerate(top5.values):
        plt.text(v + 0.5, i, str(v), va='center')
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.show()

top5_produtos_mais_vendidos(loja, 'Loja 1')
top5_produtos_mais_vendidos(loja2, 'Loja 2')
top5_produtos_mais_vendidos(loja3, 'Loja 3')
top5_produtos_mais_vendidos(loja4, 'Loja 4')
```

```python
# Função para plotar o Top 5 produtos menos vendidos
def top5_produtos_mais_vendidos(df, nome_loja):
    top5 = df['Produto'].value_counts().tail(5)

    plt.figure(figsize=(8,5))
    plt.barh(top5.index, top5.values, color='mediumseagreen')
    plt.title(f'Top 5 Produtos Mais Vendidos - {nome_loja}')
    plt.xlabel('Quantidade Vendida')
    plt.ylabel('Produto')
    plt.gca().invert_yaxis()
    for i, v in enumerate(top5.values):
        plt.text(v + 0.5, i, str(v), va='center')
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.show()

top5_produtos_mais_vendidos(loja, 'Loja 1')
top5_produtos_mais_vendidos(loja2, 'Loja 2')
top5_produtos_mais_vendidos(loja3, 'Loja 3')
top5_produtos_mais_vendidos(loja4, 'Loja 4')
```

# 5. Frete Médio por Loja


```python
def frete_medio(df):
    return df['Frete'].mean()

# Cálculo da média de frete por loja
fretes = {
    'Loja 1': frete_medio(loja),
    'Loja 2': frete_medio(loja2),
    'Loja 3': frete_medio(loja3),
    'Loja 4': frete_medio(loja4)
}

for loja, frete in fretes.items():
    print(f'{loja}: R$ {frete:.2f}')
```

```python
lojas = list(fretes.keys())
fretes_medios = list(fretes.values())

plt.figure(figsize=(8,5))
plt.bar(lojas, fretes_medios, color='salmon')
plt.title('Frete Médio por Loja')
plt.xlabel('Lojas')
plt.ylabel('Frete Médio (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, v in enumerate(fretes_medios):
    plt.text(i, v + 1, f'R$ {v:.2f}', ha='center', fontweight='bold')

plt.show()
```
